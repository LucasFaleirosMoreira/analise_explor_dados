```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import statistics as est
```

### **Realizando a Analise de dados exploratória**

```python
#Primeiro fazemos a importação dos dados e visualizamos as dimensões dos dados
base = pd.read_csv("/dados.csv")
base.shape
```

```python
#Visualizamos uma amostra para ver se está tudo correto e sabermos quais colunas devem ser usadas para analisar
base.head()
```

```python
#Depois vamos ver a linha de densidade das colunas numéricas

#visualizando a linha de densidade da coluna PIB
srn.histplot(base.PIB, kde=True, bins = 10).set(title='PIB')
```

```python
#visualizando a linha de densidade da coluna VALOREMPENHO
srn.histplot(base.VALOREMPENHO, kde=True, bins = 10).set(title='Empenho')
```

```python
#E então começamos a fazer a visualização da distribuição de valores e para cada municipio

#utilizamos a função sort_values da biblioteca pandas para ordenar de forma crescente os dados da coluna PIB de forma crescente,
#e depois com o comando head(10) escolhemos os 10 primeiros valores (ou seja, os 10 menores valores dos maiores valores)
agrupado = base.sort_values('PIB').head(10)

#e entao fazemos a variavel receber apenas os dados localizados na coluna 1 (Municipio) até a coluna anterior a coluna 3 (PIB) pegando todas as linhas (definido pelo ':')
agrupado = agrupado.iloc[:, 1:3]

#utilizamos o plot bar da biblioteca pandas para gerar um gráfico de barras verticais com a coluna "MUNICIPIO" sendo o eixo X e "PIB" o eixo Y
agrupado.plot.bar(x = 'MUNICIPIO', y = 'PIB', color = 'gray')
```

```python
agrupado2 = base.sort_values('VALOREMPENHO').head(10)

agrupado2 = agrupado2.iloc[:, 1:4]

agrupado2.plot.bar(x = 'MUNICIPIO', y = 'VALOREMPENHO', color = 'gray')
```

Com a análise dos dados fornecidos, percebemos que poucos munícipios possuem PIB maiores do que 50.000 e que o valor de empenho não parece acompanhar o PIB, pois os munícipios com maior PIB não possuem maior Empenho.
